## Row Filterings

```python
# Find top rating
```

```python
top_rated=df[df["rating"]>4.0]
```

```python
print(top_rated)
```

```python
and_condition_demo=df[(df["rating"]>4.0) & (df["discount"]>50.0)]
print(and_condition_demo)
```

```python
# Keep rows where the category is either backpacks or shoes
specific_categories = df[df['category'].isin(['backpacks', 'casual-shoes'])]
print(specific_categories)
```

```python
dicounted_items_only=df[df["discount"].notnull()]
print(dicounted_items_only)
```

```python
# find rows with missing disccount
missing_discount=df[df["discount"].isnull()]
print(missing_discount)
```
